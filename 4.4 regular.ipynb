```python
a = 'my phone is 134-1234-1234'
re.search('\d{3}-\d{4}-\d{4}', a)
b= 'my phone is 13412341234'
re.search('\d{3}\d{4}\d{4}',b)
re.search('\d{3}-{0,1}\d{4}-{0,1}\d{4}',b)
re.search('\d{3}-?\d{4}-?\d{4}',b)
```

```python
import re
email = 'david@largitdata.com'
re.match('\w+@\w+',email)
re.match('\w+@[a-z\.]+',email)
m = re.match('(\w+)@([a-z\.]+)',email)
m.group(1)
m.groups()
n = re.match('(\w+)@(.+)',email)
n.group(2)
```

```python
digit = '1999.5'
m = re.match('(\d+)\.(\d+)',digit)
m.groups()
```

#### ?P<>类型于取别名

```python
name = 'David Chiu'
m = re.match('(\w+) (\w+)',name)
m.group(2)
n = re.match('(?P<first_name>\w+) (?P<last_name>\w+)',name)
n.group('first_name')
```

```python
str1 = 'scp file.txt root@10.0.0.1:./'
m = re.match('^scp ([\w\.]+) (\w+)@([\d\.]+):(.+)',str1)
if m:
    print(m.group(1),m.group(2),m.group(3),m.group(4))
```
