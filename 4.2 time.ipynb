```python
from datetime import datetime
curent_time = datetime.now()
```

```python
curent_time
```

```python
type(curent_time)
```

## 转换成字符串

```python
datetime.strftime(curent_time,'%Y-%m-%d')
```

## 转换成时间

```python
datetime.strptime('2023/07-03','%Y/%m-%d')
```

## 时间回溯

```python
from datetime import timedelta
newtime = curent_time - timedelta(days=2)
```

```python
newtime
```

```python
newtime-curent_time
```

```python
for i in range(1,10):
    d = curent_time + timedelta(days=i)
    print(d.strftime('%Y/%m/%d'))
```

```python
from time import mktime
mktime(curent_time.timetuple())
```
