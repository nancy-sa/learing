```python
import pandas
df = pandas.read_csv("Data/Region_Data.csv",encoding='gb2312',skiprows=3,skipfooter=2,engine='python')
df.head()
df.tail()
```

#### col_level为0，则所有列名都变成行

```python
df = pandas.melt(df, col_level=0, id_vars='地区')
df
```

```python
df['variable'] = df['variable'].map(lambda e:int(e.split('年')[0]))
df.head()
```

```python
df.columns = ['area', 'year', 'gross_product']
df
```

#### 存储到数据库

```python
import pandas as pd
from sqlalchemy import create_engine
host = 'localhost'
port = 3306
username = 'root'
password = '123456'
database = 'test01'
# 建立与MySQL数据库的连接
engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}:{port}/{database}')

# 将DataFrame存储到MySQL数据库中
table_name = 'region'
df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)

# 关闭数据库连接
engine.dispose()
```

```python
import pandas
import pymysql
with pymysql.connect(host='localhost',user='root',port=3306,password='123456',charset='utf8',database='test01') as db:
    df3 = pandas.read_sql('SELECT * FROM region WHERE year = 2015 ORDER BY gross_product DESC LIMIT 3', con = db)
df3.head()
```
