#### 源码是动态数据，结合selenium来进行解析

```python
from selenium import webdriver
from bs4 import BeautifulSoup
browser = webdriver.Chrome()
browser.get('https://new.qq.com/')
html_doc = browser.page_source
soup = BeautifulSoup(html_doc,'html.parser')
element = soup.select('.itme-ls h3 a')
newsarry = []
for line in element:
    newsarry.append({'title':line.text,'url':line['href']})
newsarry
```

```python
import pandas 
newsdf = pandas.DataFrame(newsarry)
newsdf
```

```python
newsdf.to_excel('news.xlsx')
```
