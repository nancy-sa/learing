```python
import pandas
df = pandas.read_excel('Data/house_sample.xlsx')
df.head()
```

### 生成虚拟变量

```python
pandas.get_dummies(df['朝向'])
```

```python
df = pandas.concat([df,pandas.get_dummies(df['朝向'])],axis=1)
```

```python
df
```

```python
df = df.drop('朝向',axis=1)
df
```
