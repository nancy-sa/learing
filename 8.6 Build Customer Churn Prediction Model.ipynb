```python
import pandas
# index_col=0是将第一列作为索引
# header=0是将第一行作为列名
df = pandas.read_csv('Data/customer_churn.csv', index_col=0, header = 0)
df.head()
```

```python
df.columns
```

```python
df = df.iloc[:,3:]
df
```

```python
df.info()
```

#### 训练模型的数据都要是数字

```python
col_values = ['international_plan','voice_mail_plan','churn']
for i in col_values:
    df[i] = df[i].map(lambda e:1 if e=='yes' else 0)
df
```

```python
# y为chun,是否流失
# x是前面所有影响chun的变量
y = df.iloc[:,-1]
X = df.iloc[:,:-1]
```

```python
y.head()
X.head()
```

### 使用决策树建立分类模型

```python
from sklearn import tree
# 调整max_depth，可改变准确率
clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(X, y)
```

```python
# 要先将.dot的文件转成.png才能显示出来
tree.export_graphviz(clf, out_file='tree.dot')
```

```python
%pylab inline
from IPython.display import Image
Image('tree.png')
```

## 检视分类结果

```python
clf.predict(X)
```

```python
import numpy
numpy.sum(y== clf.predict(X)) / len(y)
```

## 使用逻辑回归

```python
from sklearn.linear_model import LogisticRegression
clf2 = LogisticRegression()
clf2.fit(X, y)
```

```python
clf2.predict(X)
numpy.sum(y== clf2.predict(X)) / len(y)
```

## 使用SVM

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X, y)
```

```python
model.predict(X)
numpy.sum(y== model.predict(X)) / len(y)
```
